# Exercise 2.4 — Eliciting Labels (Travel Assistant)

You need to evaluate whether a travel assistant's flight recommendations respect the user's budget constraint.

**(a)** Write a direct-grading rubric with two labels: **Within Budget** and **Over Budget**, and clear criteria for each.

**(b)** Draft a pairwise-comparison instruction for an annotator to choose which of two flight suggestion lists better adheres to the budget.

**(c)** In 2–3 sentences, explain when you would prefer direct grading versus pairwise comparison for this task.

```python
from dotenv import load_dotenv
from litellm import completion
load_dotenv()
```

## Sample data

A user request and two flight recommendation lists to use in the exercises below.

```python
user_request = """Find me round-trip flights from New York (JFK) to London (LHR)
departing March 25 and returning April 2. My budget is $800 per person."""

recommendation_a = """
1. Delta DL101 — JFK → LHR Mar 25 18:30 / LHR → JFK Apr 2 10:15 — $762/person (1 stop, ATL)
2. Norse Atlantic NO202 — JFK → LHR Mar 25 22:00 / LHR → JFK Apr 2 14:30 — $489/person (nonstop)
3. United UA303 — JFK → LHR Mar 25 20:45 / LHR → JFK Apr 2 09:00 — $831/person (nonstop)
"""

recommendation_b = """
1. British Airways BA178 — JFK → LHR Mar 25 19:00 / LHR → JFK Apr 2 11:30 — $695/person (nonstop)
2. Icelandair FI614 — JFK → LHR Mar 25 21:15 / LHR → JFK Apr 2 13:00 — $580/person (1 stop, KEF)
3. TAP Portugal TP501 — JFK → LHR Mar 25 23:00 / LHR → JFK Apr 2 16:45 — $620/person (1 stop, LIS)
"""
```

## (a) Direct-grading rubric

```python
# TODO: write a prompt that applies a direct-grading rubric (Within Budget / Over Budget)
# to evaluate recommendation_a against the user's budget constraint.

prompt = f"""
  You are responsible for evaluating if all flights adhere to a user's budget. 

  Here's the user request:
  <user_request>
  {user_request}
  </user_request>
  
  Here are the recommended flights:
  <flights>
  {recommendation_a}
  </flights>

  Answer with:
  - Within Budget: All suggested flights are equal to or less than the user's budget
  - Over Budget: At least one suggested flight exceeds the user’s budget.
"""

response = completion(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)
print(response.choices[0].message.content)
```

## (b) Pairwise-comparison instruction

```python
# TODO: write a prompt that asks the model to compare recommendation_a vs recommendation_b
# and choose which one better adheres to the user's budget.

prompt = f"""
  You are given a user query: 

  <user_request>
  {user_request}
  </user_request>

  You also have two lists of three
  flight options (A and B). 

  <flights_a>
  {recommendation_a}
  </flights_a>

  <flights_b>
  {recommendation_b}
  </flights_b>

  Choose which list better respects the budget constraint and reply with “A” or“B,” followed by a one-sentence justification.
"""

response = completion(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)
print(response.choices[0].message.content)
```
