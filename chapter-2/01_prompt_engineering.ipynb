# Exercise 2.1 — Prompt Engineering

You need to summarize customer support emails into a list of action items.

**(a)** Write a zero-shot prompt including instruction and formatting constraints.

**(b)** Extend it to a one-shot prompt with an example.

```python
from dotenv import load_dotenv
from litellm import completion
load_dotenv()
```

## (a) Zero-shot prompt

```python
email = """\
From: jane.doe@gmail.com
To: support@acmecorp.com
Subject: Login issue + double charge on invoice #4521
Date: Mon, 17 Mar 2026 09:12:00 -0300

Hi Support,

I've been trying to log in to my account for the past two days but keep getting
an "invalid password" error. I've tried resetting my password three times and
never received the reset email — I've checked spam too.

On top of that, I noticed I was charged twice on my last invoice (#4521).
The duplicate charge of $49.99 appeared on March 15.

Could you please fix both issues as soon as possible?

Thanks,
Jane Doe
Account ID: AC-20948
"""

# TODO: write your zero-shot prompt
prompt = f"""
Summarize the following support email into bullet-point action items. Email:
<email>
{email}
</email>
"""

response = completion(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)
print(response.choices[0].message.content)
```

## (b) One-shot prompt

```python
example_email = """\
From: carlos.mendes@outlook.com
To: support@acmecorp.com
Subject: Cannot access premium features after upgrade
Date: Fri, 14 Mar 2026 14:35:00 -0300

Hello,

I upgraded my plan to Premium on March 12 (order #ORD-8837) but I still only
see the Free tier features. I've tried logging out and back in, clearing my
cache, and using a different browser — nothing works.

Also, the confirmation email I received says "Free Plan" even though my credit
card was charged $29.99 for the Premium upgrade.

Can someone look into this?

Best,
Carlos Mendes
Account ID: AC-31205
"""


# TODO: extend your prompt with an example using example_email and example_action_items
prompt = f"""
Summarize the following support email into bullet-point action items. Email:
<email>
  {email}
</email>

Here is an example of how to summarize an email:
<example>
<email> {example_email} </email>
<summary> 
  User: carlos.mendes@outlook.com
  Reported Issues: 
     - Upgrade did not work
     - Wrong confirmation email received
  User Actions:
     - Logged out/in
     - Cleared cache
     - Tried different browser
  User Expectations:
     - Someone to look into his problem
</summary>
"""

response = completion(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)
print(response.choices[0].message.content)
```
