## Testing with OpenAI

A quick test call to `gpt-4o-mini` to verify your OpenAI API key is working.

```python
response = completion(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": "Say hello in one sentence."}]
)
print(response.choices[0].message.content)
```
